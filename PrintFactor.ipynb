```python
# Broot force is 
num = 10
res = []
for i in range(1,num + 1):
    if num % i == 0:
        res.append(i)
print(f"number {num} factor/Divisor is {res}")
```

```python
# second aproch
num = 30

res = []
for i in range(1,num//2 + 1):
    if num % i == 0:
        res.append(i)
res.append(num)

print('Factor is ', res)
```

```python
# optimal solution 

from math import sqrt
num = 36
'''1---------25          36/1 = 36 
   2---------18          36/2 = 18
   3---------12          36/3 = 12
   4---------9           36/4 = 9
   5---------6.2 ,      5 is not Divisor
   6---------6           36/6 = 6'''


res = []
for i in range(1,int(sqrt(num))+1):
    if num % i == 0:
        res.append(i)
        if num//i != i:
            res.append(num//i)
res.sort()
print(res)
```
