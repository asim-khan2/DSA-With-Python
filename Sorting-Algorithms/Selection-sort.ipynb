Selection Sort is a sorting algorithm that works by
finding the smallest element from the unsorted part of the list
and placing it at the correct position step by step.

```python
## selection sort ascending order
def selection_sort(nums):
    l = len(nums) ## length of the list
    for i in range(l-1): ## outer  loop
        min_index = i
        for j in range(i+1,l):## inner loop
            if nums[j] < nums[min_index]:
                min_index = j
        nums[i],nums[min_index] = nums[min_index],nums[i]

nums = [2,3,5,3,4,5,2]

selection_sort(nums)
print(nums)
```

```python
## selection sort descending order
def selection_sort1(nums1):
    l = len(nums1) ## length of the list
    
    for i in range(l-1): ## n-1
        min_index = i 
        for j in range(i+1,l):
            if nums1[i] < nums1[j]:
                min_index = j
        nums1[min_index],nums1[i] = nums1[i],nums1[min_index]
nums1 = [2,3,5,7,9,2,10,1]
selection_sort1(nums1)
print(nums1)
```
