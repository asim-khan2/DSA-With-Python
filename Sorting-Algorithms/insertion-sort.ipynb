```python
def insertion_sort(arr):
    n = len(arr) ## length of the array/list
    for i in range(1,n):
        key = arr[i] ## current element
        j = i-1

        # Move elements that are greater than key
        # to one position ahead
        while j >=0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
```

```python
arr = [3,4,66,34,56,65,23,45]
insertion_sort(arr)
print(arr)
```
