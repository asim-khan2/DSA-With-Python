```python
## constraints
## 1<=n[i]<10
## 2 n can have 10^8 elements
## 3 m can have 10^8 elements

## brute force
n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

for i in m:
    count = 0
    for j in n:
        if i == j:
            count += 1
    
    if count > 0:
        print(i, "->", count)
    else:
        print(i, "Not Present")


```

```python
has_map = [0] * 11 ## create the list size is 10 in index = 11

for num in n:
    has_map[num] += 1

for num in m:
    if num < 0 or num > 10:
        print(f"{num} : {0},",end=" ")
    else:
        print(f"{num} : {has_map[num]},",end=" ")


```

```python
## Using Dictioniory
freq = {}

# Count frequency of n
for num in n:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

# Check for m
for num in m:
    if num in freq:
        print(num, "->", freq[num])
    else:
        print(num, "->", 0)
```
