#### Recursion is a process in which a function solves a problem by calling itself with smaller inputs until a base condition is met.

```python
## recursion 
## 1 Head Recursion
## first job then function call


## Global variable
count = 0
def greet():
    global count
    if count == 4:
        return 
    print("Hello")
    count += 1
    greet()

print(greet())
```

```python
## 2 Tail Recursion
## first function call then job

count = 0

def greet():
    global count
    if count == 4:
        return
    count += 1
    greet()
    print("Hello")
    
greet()
```

```python
## Recursion using perameter


count = 0 # global variable
def fun(x,n):
    global count 
    if n == 0:
        return
    count += 1
    print(x)
    fun(x,n-1)

fun(15,4)
```

```python
## print 1 to n using Recursion
## 1. method 
def num_print(num):
    if num == 0:
        return
    print(num,end=" ")
    num_print(num-1)

num_print(5)
```

```python
## method no. 2

def print_number(i,n):
    if i > n:
        return
    print(i,end=" ")
    print_number(i+1,n)

print_number(1,5)

## Time Complexity O(n)
## space Complexity O(n)
```

```python
## Using Tail Recursion
## print number N to 1
def print_number(i,n):
    if i > n:
        return
    print_number(i+1,n)
    print(i)
print_number(1,5)
```

```python
## Perameterized and function Recursion
## Exampel sum of 1 to N

## Fuction Recursion
def sum(n):
    if n == 1:
        return 1
    return n + sum(n-1)

print(sum(5))
```

```python
## sum
## perameterized recursion
def sumof(sum,i,n):
    if i>n:
        print(sum)
        return
    sumof(sum+i,i+1,n)

sumof(0,1,4)
```

```python
## Factorial 

def factorial(n):
    if n == 1 :
        return 1
    return n * factorial(n-1)
## function call
factorial(5)
```

```python
## Reverse array
nums = [2,3,5,7,8,9,4,4,6,7]
## function of the reversed 
def reversed(nums):
    ## left pointer
    left = 0
    ## right pointer
    right = len(nums)-1

    while(left < right):
        temp = nums[left]
        nums[left] = nums[right]
        nums[right] = temp
        left += 1
        right -= 1
    return nums

print(reversed(nums))
```

```python
## reversed array using recution
nums = [2,3,5,7,8,9,4,4,6,7]


def reverse(nums,left,right):
    if left >= right:
        return
    nums[left],nums[right] = nums[right],nums[left]
    reverse(nums,left+1,right-1)

print("This is the before reverse")
print(nums)
reverse(nums,0,len(nums)-1)
print("This is the after reversed")
print(nums)
```

```python
## check the string palindrom or not

def is_palindrome(s,left,right):
    if left >= right:
        return True
    if s[left] != s[right]:
        return False
    return is_palindrome(s,left+1,right-1) 

s = "racecar"

bool = is_palindrome(s,0,len(s)-1)
print(bool)
if bool == True:
    print("string is  the palindrome")
else:
    print("string is not palindrome")
```

```python
### Using while loop
def is_palindrome(s,left,right):
    while(left >= right):
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

s = 'racecar'

check = is_palindrome(s,0,len(s)-1)
if check == False:
    print("String is not palindrome")
else:
    print("String is palindrome")
    
```

```python
## Fibonacci  Series

def fibo(n):

    if n==0 or n==1:
        return n
    return fibo(n-1)+fibo(n-2)

print(fibo(5))
     
```
